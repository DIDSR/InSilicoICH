```python
import matplotlib.pyplot as plt
from insilicoICH import load_phantom
import numpy as np
age = 38
seed = 30
lesion_count = 50
verbose = False

input_volumes = np.linspace(1, 70, num=lesion_count)
print(input_volumes)
```

```python
#input_volumes = [30, 30, 30, 30, 30, 30, 30, 30, 30, 30, 30]
output_EDH_volumes = np.zeros_like(input_volumes)

#input_volumes = [input_volumes[21]]

for idx, volume in enumerate(input_volumes):
    if verbose: print(f'Inserting EDH lesion {idx+1}/{lesion_count}')
    phantom = load_phantom(age)

    phantom.insert_lesion('EDH', mass_effect=False, seed=None, intensity=80, volume=volume)
    output_EDH_volumes[idx] = phantom.get_lesion_volume()

    if volume == input_volumes[21]:
        example_EDH = phantom._phantom

        
```

```python
from utils import scrollview
scrollview(example_EDH, display='soft tissues')
```

```python
output_SDH_volumes = np.zeros_like(input_volumes)

input_volumes = np.linspace(1, 70, num=lesion_count)
# print(input_volumes)
# input_volumes = [input_volumes[24]]
# print(input_volumes)

for idx, volume in enumerate(input_volumes):
    if verbose: print(f'Inserting SDH lesion {idx+1}/{lesion_count}')
    phantom = load_phantom(age)

    phantom.insert_lesion('SDH', mass_effect=False, seed=None, intensity=80, volume=volume)
    output_SDH_volumes[idx] = phantom.get_lesion_volume()

    if volume == input_volumes[21]:
        example_SDH = phantom._phantom
```

```python
from utils import scrollview
scrollview(example_SDH, display='soft tissues')
```

```python
output_IPH_volumes = np.zeros_like(input_volumes)

for idx, volume in enumerate(input_volumes):
    if verbose: print(f'Inserting IPH lesion {idx+1}/{lesion_count}')
    phantom = load_phantom(age)

    phantom.insert_lesion('IPH', mass_effect=False, seed=None, intensity=80, volume=volume)
    output_IPH_volumes[idx] = phantom.get_lesion_volume()
```

```python
fig, axs = plt.subplots(nrows=1, ncols=3, figsize=(20,5))
axs[0].set_title('EDH Input/Output')
axs[0].scatter(input_volumes, output_EDH_volumes)
axs[0].plot(input_volumes, input_volumes, '--')
axs[0].set_xlabel('Input Volumes (mL)')
axs[0].set_ylabel('Output Volumes (mL)')

axs[1].set_title('SDH Input/Output')
axs[1].scatter(input_volumes, output_SDH_volumes)
axs[1].plot(input_volumes, input_volumes, '--')
axs[1].set_xlabel('Input Volumes (mL)')
axs[1].set_ylabel('Output Volumes (mL)')

axs[2].set_title('IPH Input/Output')
axs[2].scatter(input_volumes, output_IPH_volumes)
axs[2].plot(input_volumes, input_volumes, '--')
axs[2].set_xlabel('Input Volumes (mL)')
axs[2].set_ylabel('Output Volumes (mL)')
```
