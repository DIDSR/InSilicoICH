```python
from ipywidgets import interact
from pedsilicoICH.ground_truth_definition.phantoms import load_phantom

from utils import make_and_display_lesion, scrollview

age = 6.5
phantom = load_phantom(age=age)

def make_and_display_lesion_function(lesion_type: str='round', volume: float=1.0, mass_effect: float = 0.5, intensity: float=100, edema: float=0, seed: int=42, complexity=1, show_mask:bool=True):
    return make_and_display_lesion(phantom, lesion_type=lesion_type, volume=volume, mass_effect=mass_effect, intensity=intensity, edema=edema, seed=seed, complexity=complexity, show_mask=show_mask)

interact(make_and_display_lesion_function,
         lesion_type=['round', 'epidural', 'subdural'],
         volume=(0, 50, 0.5),
         mass_effect=(0, 1, 0.1),
         intensity=(-20, 100, 10),
         edema=(0, 20, 1),
         seed=(0, 100, 1),
         complexity = (1, 5, 1),
         show_mask=True)
```
